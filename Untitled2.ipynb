```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Feature1': [10, 15, 8, 12, 9, 7],
    'Feature2': [25, 20, 15, 22, 18, 14],
    'Feature3': [8, 10, 12, 6, 9, 11]
}

df = pd.DataFrame(data)

# Importing parallel_coordinates function from pandas.plotting module
from pandas.plotting import parallel_coordinates

# Create a parallel coordinates plot
plt.figure(figsize=(8, 5))
parallel_coordinates(df, 'Category', colormap='viridis', alpha=0.7)

# Add a legend
plt.legend(loc='upper right')

# Display the plot
plt.title('Parallel Coordinates Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Plot the data
plt.plot(x, y)

# Set custom x-axis tick locations and labels
plt.xticks([1, 3, 5], ['A', 'B', 'C'])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7, 8]

# Plotting a histogram with bins
plt.hist(data, bins=12, edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the plot
plt.show()
```
